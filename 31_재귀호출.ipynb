```python
# 재귀 호출 사용하기

def hello():
    print('Hello, world!')
    hello()
```

```python
hello() # 종료하지 않으면 최대 재귀 길이를 초과하여 에러 발생
```

```python
## 재귀 호출에 종료 조건 만들기
def hello(count):
    if count == 0:
        return
    
    print('Hello, world', count)
    
    count -= 1
    hello(count)

hello(5)
```

```python
# 재귀호출로 팩토리얼 구하기

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
factorial(5)
```
