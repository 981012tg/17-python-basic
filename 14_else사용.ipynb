```python
#if 조건문에 숫자 지정하기

#0이면 거짓, 0이 아니면 실수와 관계없이 참

if 0:
    print("참")
else:
    print("거짓")

if 1:
    print("참")
else:
    print("거짓")

if 0x1F:
    print("참")
else:
    print("거짓")

if 0b1000:
    print("참")
else:
    print("거짓")

if 13.5:
    print("참")
else:
    print("거짓")
```

```python
#조건문에 문자열 지정하기
if "hello":         # 내용이 있기만 하면 참
    print ("참")
else:
    print("거짓")

if " ":             # 공백이라도 있으면 참                     
    print ("참")
else:
    print("거짓")

if "":              # 아무것도 없으면 거짓
    print ("참")
else:
    print("거짓")
```

```python
if not 0:
    print("참")

if not None:
    print("참")

if not "":
    print("참")


#거짓을 not으로 뒤집으면 참이 됨
```

```python
# 연습문제(p.174)
#합격여부 판단하기
#시험점수가 80점 이상이면서 코딩 시험을 통과해야 합격
#코딩시험 통과 여부는 True or False
#합격 or 불합격 출력 되게 만들기

written_test = 75
coding_test = True

if written_test >= 80 and coding_test == True:
    print("합격")
else:
    print("불합격")
```

```python
#심사문제(p.174)
# 표준 입력으로 국어 영어 수학 과학 점수가 입력됨
# 네과목 평균점수가 80점 이상일때 합격
# 평균점수에 다라 합격 불합격 출력하기
# 범위는 0~100점까지인데, 범위를 벗어나면 '잘못된 점수'를 출력하고 합격, 불합격 여부는 출력하면 안됨

korean, english, math, science = map(int,(input().split()))

avg=(korean+english+math+science)/4

if korean < 0 or korean > 100 or english < 0 or english > 100 or math < 0 or math > 100 or science < 0 or science > 100:
    print('잘못된 점수')

else:
    if avg >= 80:
        print("합격")
    else:
        print("불합격")
```
