```python
#변수만들기 
## 1 영문문자와 숫자를 사용
## 2 대소문자를 구분
## 3 문자부터 시작해야하고, 숫자로 시작하면 안됨 (ex. 1a=10 은 안됨)
## 4 _로 시작할 수 있음
## 5 특수문자(!, @, #, $, %, ^, &, *, (, ),-,+ 등)는 사용할 수 없음
## 6 파이썬의 키워드(if, for, while 등)는 변수로 사용할 수 없음


#convention
# stand_alone=1  이런식으로 언더스코어를 많이씀
# StandAlone=2   이건 다른언어에서 쓰긴 하는데 파이썬에서는 잘 안씀
# standAlone=3   이런식으로 카멜케이스도 있긴 한데, 권장하진 않는다
# stand-alone=4   이런식으로 특수문자를 쓰면 안됨
# stand alone=5   이런식으로 띄어쓰기도 안됨
```

```python
# 변수 여러개를 한번에 만들기
x, y, z = 10, 20, 30
print(x, y, z)
print(x)
print(y)
print(z)

#x, y, z = 10, 20  이런식으로 일대일로 대응이 안되면 안됨
```

```python
#a,b= int(input("숫자 두개를 입력하세요")).split()

#이렇게 하면 안됨. 왜? 10 20 은 문자열이기 때문에 int로 변환이 안됨. 그래서 split()으로 나눈 후 각각 int로 변환해야함
```

```python
a,b= input("숫자 두개를 입력하세요").split()
a=int(a)  #문자열 10이 정수 10으로 바뀐거고
b=int(b)  #문자열 20이 정수 20으로 바뀐거임
print(a)
print(b)
```

```python
print(map(int, ["10", "20"]))  #이렇게 하면 아직 memory에 할당된 상태
print(list(map(int, ["10", "20"])))  #이렇게 하면 list로 변환됨
```

```python
a,b=map(int,input("숫자 두개를 입력하세요").split())
print(a)
print(b)
print(a+b)
```

```python
#입력받은 값을 콤마를 기준으로 분리하기
a,b=map(int,input("숫자 두개를 입력하세요").split(","))
print(a)
print(b)
print(a+b)
```

```python
print("10,20".split())
print("10,20".split(","))
print("10,2,0".split(","))
```
