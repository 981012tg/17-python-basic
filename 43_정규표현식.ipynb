```python
#문자열 판단하기

import re
re.match('Hello', 'Hello, world!')
```

```python
re.search("^hello", "hello, world")
```

```python
re.search("world!$","hello, world!")
```

```python
## 지정된 문자열이 하나라도 포함되는지 판단하기

re.match('hello|world', 'hello')
```

```python
# 범위판단하기

print(re.match('[0-9]*', '1234'))
print(re.match('[0-9]+', '1234'))
print(re.match('[0-9]+', 'abcd'))
```

```python
print(re.match('a*b', 'b'))
print(re.match('a+b', 'b'))
print(re.match('a*b', 'aab'))
print(re.match('a+b', 'aab'))
```

```python
#문자가 한개만 있는지 판단하기
# 문자?
# [0-9]?

print(re.match("abc?d", "abd"))
print(re.match("ab[0-9]?c", "ab3d"))
print(re.match("ab.d", "abxd"))
```

```python
## 문자 개수 판단하기
# 문자(개수)
# (문자열)(개수)

print(re.match("h{3}", "hhhello"))
print(re.match("(hello){3}", "hellohellohelloworld"))
```

```python
# [0-9](개수)
re.match("[0-9]{3}-[0-9]{4}-[0-9]{4}", "010-0000-0000")
```

```python
## 숫자와 영문 문자를 조합해서 판단하기

# a-z
# A-z

print(re.match('[a-zA-Z0-9]+', 'Helo1234'))
print(re.match("[A-Z0-9]+", "hello"))
```

```python
re.match('[가-힣]+', "홍길동")  #가 = 제일 첫 한글, 힣 = 제일 마지막 한글
```

```python
## 특정 문자 범위에 포함되지 않는지 판단하기

# [^범위]*
# [^범위]+

print(re.match('[^A-Z]+', 'Hello'))
print(re.match('[^A-Z]+', 'hello'))
```

```python
# [^범위]*
# [^범위]+
print(re.match('^[A-Z]+', 'Hello'))
```

```python
# [범위]*$
# [범위]+$

re.search('[0-9]+$', 'Hello1234')
```

```python
## 특수문자 판단하기

print(re.search("\*+", "1 ** 2"))
```

```python
re.search('[$()a-zA-Z0-9]', '$(document)')
```

```python
# \d -> [0-9]
# \D -> [^0-9]
# \w -> [a-ZA-Z0-9_]
# \W -> [^a-ZA-Z0-9_]

print(re.match("\\d+", "1234"))
print(re.match("\\D+", "Hello"))
print(re.match("\\w+", "Hello_1234"))
print(re.match("\\W+", "(:)"))
```

```python
## 공백처리하기

print(re.match("[a-zA-Z0-9 ]+", "Hello 1234"))
print(re.match("[a-zA-Z0-9]+", "Hello 1234"))
print(re.match("[a-zA-Z0-9\\s]+", "Hello 1234"))
```

```python
# 그룹사용하기

# (정규표현식)(정규표현식)
# 매체객체.group(그룹숫자)

m = re.match('([0-9]+) ([0-9]+)', '10 295')

print(m.group(1))
print(m.group(2))
print(m.group())
print(m.group(0))
```

```python
print(m.groups())
print(m.groups()[0])
```

```python
# (?P<이름>정규표현식)

m = re.match('(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\\((?P<arg>\\w+)\\)', 'print(1234)')

print(m.group('func'))
print(m.group('arg'))
```

```python
## 패턴에 매칭되는 모든 문자열 가져오기

# re.findall('패턴','문자열')

re.findall('[0-9]+', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
#문자열 바꾸기
# re.sub("패턴", "바꿀문자열", "문자열", 바꿀횟수)

re.sub("apple|orange", "fruit", "apple box orange tree")
```

```python
re.sub('[0-9]+', 'n', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
# 교체함수(매치객체)
# re.sub('패턴', 교체함수, '문자열', 바꿀횟수)

def multiple10(m):
    n = int(m.group())
    return str(n * 10)

re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
re.sub('[0-9]+', lambda m: str(int(m.group())*10), '1 2 Fizz 4 Buzz Fizz 7 8')
```
