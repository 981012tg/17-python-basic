```python
#딕셔너리 조작

## setdefault, update

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}
```

```python
a.setdefault("e")
print(a)
```

```python
a.setdefault("f", 100)
print(a)
```

```python
## 딕셔너리의 키의 값 수정하기

a = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}
```

```python
a.update(a=90)
print(a)
```

```python
a.update(e=50)
print(a)
```

```python
a.update(a=900, f=60)
print(a)
```

```python
y = {
    1 : 'one',
    2 : 'two'
}

y.update({1:'ONE', 2:'TWO'})

print(y)
```

```python
y.update(
    [
        [2, 'TWO'],
        [4, 'FOUR']
        ]
)

print(y)
```

```python
y.update(
    zip(
        [1, 2],
        ['one', 'two']
	)
)

print(y)
```

```python
## 딕셔너리 표현식(comprehension) 사용하기

# {키 : 값 for 키, 값 in 딕셔너리}
# dict({키 : 값 for 키, 값 in 딕셔너리})

keys = ['a', 'b', 'c', 'd']
imsi = {
    'a' : None,
    'b' : None,
    'c' : None,
    'd' : None,
}

x={key : value for key, value in imsi.items()}
print(x)
#{key : value for key, value in dict.fromkeys(keys).items()}
```

```python
{key : 0 for key in dict.fromkeys(['a', 'b', 'c', 'd']).keys()}
```

```python
{value : 0 for value in dict.fromkeys(['10', '20', '30', '40']).keys()}
```

```python
{value : 0 for value in {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}.values()
}
```

```python
## 딕셔너리 표현식에서 if 조건문 사용하기

x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

for key, value in x.items():
    if value == 20:
        del x[key]

print(x)
```

```python
#위에껀 값을 지우는거였는데, 이럴경우, 반복 도중 딕셔너리의 크기가 바뀌어서 에러가남
#지금 할건, 지우는게 아니라, 제외할 값을 뺀 나머지 애들로 새로운 딕셔너리를 만드는 방법


x = {
    "a" : 10,
    "b" : 20,
    "c" : 30,
    "d" : 40
}

x={key : value for key, value in x.items() if value !=20} 
print(x)
```
