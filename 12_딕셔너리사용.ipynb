```python
# 딕셔너리 만들기

# 딕셔너리 = {
#     키1 : 값1,
#     키2 : 값2,
#     ...
# }
```

```python
lux ={
    "health" : 490,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}
```

```python
#키 이름이 중복되면?
lux ={
    "health" : 490,
    "health" : 800,
    "mana" : 334,
    "melee" : 550,
    "armor" : 18.72
}

print(lux) #키가 중복되면 가장 뒤에 있는 값만 사용함
```

```python
#딕셔너리 키의 자료형
x={
    100: 'hundred',
    False : 0,
    3.5: [3.5, 3.5]
}

print(x)

#값에는 모든 자료형을 쓸 수 있다. 단, 키에는 리스트와 딕셔너리를 쓸 수 없음
```

```python
# 빈 딕셔너리 만들기
x={}
y=dict()

print(x)
print(type(x))

print(y)
print(type(y))
```

```python
#dict로 딕셔너리 만들기

lux1=dict(
    health=490,   #이때는 '', ""를 쓰면 안됨
    mana=334,
    melee=50,
    armor=18.72 
)
print(lux1)
```

```python
lux2=dict(zip(
    ['health', 'mana', 'melee', 'armor'],
    [490, 334, 50, 18.72]
))

print(lux2)
```

```python
lux3=dict([
    ('health', 490),
    ('mana', 334),
    ('melee', 50),
    ('armor', 18.72)
])
print(lux3)
```

```python
lux4=dict({'health': 490, 'mana': 334, 'melee': 50, 'armor': 18.72})

print(lux4)

#이건 굳이..?
```
