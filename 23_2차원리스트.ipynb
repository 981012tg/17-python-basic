```python
col, row = map(int, input().split())

matrix = []

for i in range(row):
	matrix.append(list(input()))

for r in range(row):
	for c in range(col):
		if matrix[r][c] == "*":
			continue

		count = 0

		for nr in range(r - 1, r + 2):
			for nc in range(c - 1, c + 2):
				if 0 <= nr < row and 0 <= nc < col:
					if matrix[nr][nc] == "*":
						count += 1

		matrix[r][c] = str(count)

print(matrix)

#for r in range(row):
#    print("".join(matrix[r]))
```

```python
korean, english, math, science = map(int, input().split())

def get_min_max_score(*args):
	return min(args), max(args)

def get_average(*args):
	return sum(args)/len(args)

min_score, max_score = get_min_max_score(korean, english, math, science)
average_score = get_average(korean, english, math, science)

print(f"낮은점수: {min_score:.2f}, 높은점수: {max_score:.2f}, 평균점수: {average_score:.2f} ")
```

```python
#문제 1

students = {

	"김철수": [90, 80, 70],
	
	"이영희": [100, 95, 90],
	
	"박민수": [60, 70, 65]

}

total_score = []

print("===== 학생 성적 결과 =====")

for name, scores in students.items():
	avg = sum(scores)/len(scores)
	print("이름: ", name)
	print("점수: ", scores)
	print("평균: ", avg)
	print()
	total_score.extend(scores)
	
print("===== 전체결과 =====")

print("전체 학생 수: ", len(students))
print("전체 평균: ", sum(total_score)/len(total_score))
	

# 출력결과 
# ===== 학생 성적 결과 =====

# 이름: 김철수

# 점수: [90, 80, 70]

# 평균: 80.0



# 이름: 이영희

# 점수: [100, 95, 90]

# 평균: 95.0



# 이름: 박민수

# 점수: [60, 70, 65]

# 평균: 65.0

# ===== 전체 결과 =====

# 전체 학생 수: 3명

# 전체 평균: 80.0
```

```python
# 문제 5.

# for문을 이용하여 학생 정보 5명을 입력받고 출력하는 프로그램을 작성하세요.

# 입력항목 
# 이름
# 나이
# 학년

# 요구사항

# 1. for문을 사용하세요.
# 2. 학생 정보는 딕셔너리로 저장하세요.
# 3. 학생 5명의 정보는 리스트에 저장하세요.
# 4. 입력이 끝나면 전체 학생 정보를 출력하세요.

students = []

for i in range(5):
	name = input("이름: ")
	age = input("나이: ")
	grade = input("학년: ")
	student = {
		"이름" : name,
		"나이" : age,
		"학년" : grade
	}
	students.append(student)

print(students)
```

```python
col, row = map(int, input().split())

matrix = []

for i in range(row):
	matrix.append(list(input()))

for r in range(row):
	for c in range(col):
		if matrix[r][c] == "*":
			continue
		
		count=0

		for nr in range(r-1, r+2):
			for nc in range(c-1, c+2):
				if 0 <= nr < row and 0 <= nc < col:
					if matrix[nr][nc] == "*":
						count += 1
		matrix[r][c] = str(count)

for r in range(row):
	print(''.join(matrix[r]))
```
