```python
# for 반복문을 두번 사용하기
a= [
	[10, 20],
	[30, 40],
	[50, 60],
]

for i in a:
	for j in i:
		print(j, end=' ')
	print() 
```

```python
# for와 range 사용하기
a= [
	[10, 20],
	[30, 40],
	[50, 60],
]

for i in range(len(a)): #len(a) == 3, 6이아니다
	for j in range(len(a[i])):
		print(a[i][j], end=' ')
	print()
```

```python
# while 반복문을 한 번 사용하기
a= [
	[10, 20],
	[30, 40],
	[50, 60],
]

i = 0
while i < len(a) :
	x, y = a[i]
	print(x, y)
	i +=1
```

```python
## 2차원 리스트 만들기

a=[]

for i in range(3):
	line = []
	for j in range(2):
		line.append(0)
	a.append(line)

print(a)
```

```python
#리스트 표현식으로 2차원 리스트 만들기

a = [[0 for j in range(2)] for i in range(3)]
print(a)
```

```python
# 톱니형 리스트 만들기

a = [3, 1, 3, 2, 5]
b= []

for i in a:
	line = []
	for j in range(i):          #for _ in range(i): 이렇게 써도 될듯
		line.append(0)
	b.append(line)

print(b)
```

```python
col, row = map(int, input().split())

matrix = []

for i in range(row):
    matrix.append(list(input()))

for r in range(row):
    for c in range(col):
        if matrix[r][c] == "*":
            continue

        count = 0

        for nr in range(r - 1, r + 2):
            for nc in range(c - 1, c + 2):
                if 0 <= nr < row and 0 <= nc < col:
                    if matrix[nr][nc] == "*":
                        count += 1

        matrix[r][c] = str(count)

#print(matrix)

for r in range(row):
    print("".join(matrix[r]))
```
