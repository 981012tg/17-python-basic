```python
#인덱스와 요소를 함께 출력하기 - enumerate

a = [38, 21, 53, 62, 19]

for index, value in enumerate(a):
    print(f"{index+1}번째 값은 {value} 입니다.")
```

```python
# while 반복문으로 요소 출력하기
a = [38, 21, 53, 62, 19]

i = 0

while i < len(a):
    print(a[i])
    i += 1
```

```python
# 리스트의 가장 작은수, 가장 큰수, 합계 구하기
# min(), max(), sum()

a = [38, 21, 53, 62, 19]
smallest = a[0]

for i in a:
    if i < smallest:
        smallest = i

print(smallest)
```

```python
a = [38, 21, 53, 62, 19]
largest = a[0]

for i in a:
    if i > largest:
        largest = i

print(largest)
```

```python
a = [38, 21, 53, 62, 19]

x = 0

for i in a:
    x += i

print(x)
```

```python
# 리스트 표현식 사용하기
# [식 for 변수 in 리스트]   이걸 많이 씀
# list(식 for 변수 in 리스트)

a = [i for i in range(10)] #변수에 맞는 식을 써야한다
print(a)

b = list(i for i in range(10))
print(b)
```

```python
c = [i + 5 for i in range(10)]
print(c)

d = [i*2 for i in range(10)]
print(d)
```

```python
# 리스트 표현식에서 if조건문 사용하기

#[식 for 변수 in 리스트 if조건식]  이걸 많이씀

#list(식 for 변수 in 리스트 if조건식)

a = [i for i in range(10) if i % 2 == 0]
print(a)
```

```python
#위에껄 풀어쓰면
a=[]
for i in range(10):
    if i % 2 == 0:
        a.append(i)

print(a)
```

```python
a = [i + 5 for i in range(10) if i % 2 == 1]
print(a)
```

```python
# for 반복문과 if 조건문을 여러번 사용하기
a = [i * j for i in range(2,10) for j in range(1,10)]
print(a)
```

```python
b=[]
for i in range(2,10):
    for j in range(1,10):
        b.append(i*j)
    

print(b)
```

```python
# 리스트에 map 사용하기

#list(map(함수,리스트))
#tuple(map(함수,튜플))

a = [1.2, 2.5, 3.7, 4.6]
for i in range(len(a)):
    a[i] = int(a[i])

print(a)
```
