```python
#값을 여러개 출력하기
print(1,2,3)
print("Hello","Python")
#이렇게 하면 각 요소 사이에 공백이 들어감
```

```python
#sep로 값 사이에 문자 넣기
print(1,2,3,sep=", ")
print(1,2,3,sep='')
```

```python
print(1920, 1000, sep='x')  #1920x1000
```

```python
#줄바꿈 활용하기
print(1,2,3, sep="\n")
```

```python
print("1\n2\n3")  #이렇게도 가능
```
