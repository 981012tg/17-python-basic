```python
# # elif 사용하기

# if 조건식:
#     코드1
# elif 조건식:
#     코드2

x=20
if x == 10:
    print("10입니다.")
elif x == 20:
    print("20입니다.")
```

```python
# if, elif, else 모두 사용
# if 조건식:
#   코드1
# elif 조건식:
#   코드2
#else:
#   코드3

x=30
if x == 10:
    print("10입니다.")
elif x == 20:
    print("20입니다.")
else:
    print("10도 20도 아닙니다.")
```

```python
#음료수 자판기 만들기

button = int(input())

if button == 1:
    print("콜라")
elif button == 2:
    print("사이다")
elif button == 3:
    print("환타")
else:
    print("제공하지 않는 메뉴")
```

```python
# 연습문제(p.180)

x = int(input())

if 11 <= x <= 20:
    print("11~20")
elif 21 <= x <= 30:
    print("21~30")
else:
    print("아무것도 해당하지 않음") 
```

```python
#심사문제(p.180)

#표준입력으로 나이가 입력된다(입력값은 7이상)
#각 나이에 맞게 요금을 차감한 뒤에 잔액이 출력되게 만들기. 현재 교통카드에는 9000원 들어있음
#어린이:초등학생, 7~12세, 650원
#청소년:중고등학생, 13~18세, 1050원
#어른:일반, 19세~, 1250원

age=int(input())
balance=9000

if 7 <= age <= 12:
    balance=balance-650

elif 13 <= age <= 18:
    balance=balance-1050

elif age >= 19:
    balance=balance-1250

else:
    print("무료입니다")

print(balance)
```
