```python
# 반복문으로 세트의 요소를 모두 출력하기
a={1,2,3,4}

for i in a:
    print(i)
```

```python
# 세트 표현식(comprehension) 사용하기
#{식 for 변수 in 반복가능한객체}
#set(식 for 변수 in 반복가능한객체)

a = {i for i in 'apple'}
print(a)
```

```python
## 세트 표현식에 if 조건문 사용하기
# {식 for 변수 in 세트 if 조건식}
# set(식 for 변수 in 세트 if 조건식)

a = {i for i in 'pineapple' if i  not in 'apl'}

print(a)
```

```python
#연습문제
a = {i for i in range(1, 101) if i % 3 == 0 and i % 5 == 0}

print(a)
```
