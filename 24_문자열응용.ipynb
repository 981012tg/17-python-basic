```python
# 문자열 서식 지정자와 포매팅 사용하기

##서식 지정자로 문자열 넣기 (옛날에 주로 쓰던 방법)
# '%s' % '문자열'

'I am %s.' % 'james'
```

```python
name = "maria"
'I am %s.' %name
```

```python
## 서식지정자로 숫자 넣기
# '%d' % 숫자

'I am %d years old.' % 20
```

```python
## 서식지정자로 소숫점 표현하기
# '%f' % 숫자

'%f' % 2.3
```

```python
#'%.자릿수f' % 숫자

'%.10f' % 2.3
```

```python
'%10.3f' % 2.3
```

```python
## 서식지정자로 문자열 정렬하기
# %길이s

'%10s' % 'python'
```

```python
'%-10s' % 'python'
```

```python
## 서식 지정자로 문자열 안에 값 여러개 넣기

#'%d %s' % (숫자, 문자열)

"Today is %d %s." % (3, 'April')
```

```python
## format 메서드 사용하기
"Hello, {0}".format('world!')
```

```python
"Hello, {0} {2} {1}".format('Python', 'Script', 3.6)
```

```python
'{0} {0} {1} {1}'.format('Python', 'Script')
```

```python
"Hello, {language} {version}".format(language='Python', version=3.6)
```

```python
## format 매서드로 문자열 정렬하기

"{0:<10}".format("python") #화살표 왼쪽 정렬
```

```python
"{0:>10}".format("python") #화살표 오른쪽 정렬
```

```python
"{:>10}".format("python") #화살표 오른쪽 정렬
```
