```python
#3의 배수일때와 5의 배수일때 처리하기
for i in range(1, 101):
    
    if i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

```python
#3과 5의 공배수 처리하기
for i in range(1, 101):
    
    if i % 3 == 0 and i % 5 == 0:   #    if i % 15로 해도 되는데 더 복잡해지면 논리연산자가 더 이해하기 쉬움
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```
