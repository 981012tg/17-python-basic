```python
for i in range(5):
    for j in range(5):
       # print("*") #5*5=25개의 별이 찍힌다
        print("j:", j, sep='', end=' ')  
    print('i:', i, '\n', sep='')
    #print()
```

```python
# 사각형으로 별 출력하기
for i in range(5):
    for j in range(5):
        print("*", end='')
    print()
```

```python
for i in range(3):
    for j in range(7):
        print("*", end='')
    print()
```

```python
# 계단식으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j <= i:
            print('*', end='')
    print()
```

```python
# 대각선으로 별 출력하기
for i in range(5):
    for j in range(5):
        if j == i:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

```python
#연습문제(p.218)
for i in range(5):
    for j in range(5):
        if j<i:
            print(' ', end='')
        else:
            print('*', end='')
    print()
```

```python
#심사문제(p.219)
height=int(input())

for i in range(height):
    for j in range(height-i-1):
        print(" ", end="")
    for k in range(2*i+1):
        print("*", end="")
    print()
```

```python
height = 5

for i in range(1, height+1):
    space = height-i
    star = 2 * i - 1
    print(" " * space + "*" * star) 
```

```python
#심사문제(p.219)
height=int(input())

for i in range(height):
    for j in range(2*i+1):
        print(" ", end="")
    for j in range(height-i-1):
        print("*", end="")
    print()
```
